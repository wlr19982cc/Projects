## Prepare Data

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq

x_10 = np.linspace(0, 2*math.pi, 10)
x_15 = np.linspace(0, 2*math.pi, 15)
x_100 = np.linspace(0, 2*math.pi, 100)

y_10 = np.sin(x_10)
y_15 = np.sin(x_15)
y_100 = np.sin(x_100)

y_noise_10 = [np.random.normal(0, 0.2) + i for i in y_10]
y_noise_15 = [np.random.normal(0, 0.2) + i for i in y_15]
y_noise_100 = [np.random.normal(0, 0.2) + i for i in y_100]
```

## Fitting Curve

```python
def fun(p, x, y):
    return y - np.poly1d(p)(x)

def fun_norm(p, x, y):
    reg = y - np.poly1d(p)(x)
    return np.append(reg, p)

curve_0 = leastsq(fun, np.random.randn(3+1), args=(x_10, y_noise_10))
curve_1 = leastsq(fun, np.random.randn(9+1), args=(x_10, y_noise_10))
curve_2 = leastsq(fun, np.random.randn(9+1), args=(x_15, y_noise_15))
curve_3 = leastsq(fun, np.random.randn(9+1), args=(x_100, y_noise_100))
curve_4 = leastsq(fun_norm, np.random.randn(9+1), args=(x_10, y_noise_10))
```

## 10 Sample and Degree 3

```python
x = np.linspace(0, 2*math.pi)
y = np.sin(x)

fig1 = plt.figure("fig1")
plt.ylim(-2, 2)
plt.xlim(-1, 2*math.pi+1)
plt.plot(x, y, color="green", linewidth="2", label="Sin(x)")
plt.plot(x, np.poly1d(curve_0[0])(x), color="red", linewidth="2", label="Fitted Curve")
plt.plot(x_10, y_noise_10 ,"o", color="blue", label="Noisy Samples")
plt.title("10 Smples and Degree 3")
plt.legend()
plt.savefig("fig1.png")
```
